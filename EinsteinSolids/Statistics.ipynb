**Computing standard deviations:**

Given some probability function, $p(x)$, the standard deviation of $x$ is given by
$$\delta x = \sqrt{\overline{x^2} - {\overline{x}}^2},$$
where $\overline{x}$ represents the *average* of $x$.  If $p(x)$ is normalized, then the average is given by
$$\overline{x}=\sum_{i}p_i(x)\,x_i.$$

In the cells below, we will create and analyze a **Gaussian distribution**. The width of this distribution is defined by the parameter `w`.

```python
from pylab import *
%matplotlib inline
```

```python
x = linspace(0, 10, 101)  # Array of x values
x0 = 5                    # Center of distribution
w = 1.0                   # Width of distribution
p = exp(-(x-x0)**2 / (2*w**2))
plot(x, p)
grid(True)
show()
```

```python
sum(p) # Note, this distribution is NOT NORMALIZED.
```

```python
p = p / sum(p) # This NORMALIZES the distribution.
sum(p)
```

```python
plot(x, p)
grid(True)
show()
```

The **AVERAGE** of the distribution:

```python
avg_x = sum(p*x)
avg_x**2
```

The **AVERAGE OF THE SQUARE** of the distribution:

```python
avg_x2 = sum(p*x**2)
avg_x2
```

The **STANDARD DEVIATION** of the distribution:

```python
sqrt(avg_x2 - avg_x**2)
```
